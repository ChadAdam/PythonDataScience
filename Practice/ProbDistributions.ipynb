```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Binomial Distribution


# N = 10 , P = 30%
binomial = np.random.binomial(10,.3, 100)
print('The Mean is :{} , and the STD is :{}'.format(binomial.mean(), binomial.std()))
plt.hist(binomial)
plt.axvline(binomial.mean(),color ='r', linewidth='2')
plt.axvline(binomial.mean()+binomial.std(), color='y', linestyle='dashed')
plt.axvline(binomial.mean()-binomial.std(), color='y', linestyle='dashed')
plt.show()
```

```python
# Gamma Distribution



# Shape =5 and Scale = 1
gamma = np.random.gamma(5,1, 1000)
print('The Mean is :{} , and the STD is :{}'.format(gamma.mean(), gamma.std()))
plt.hist(gamma)
plt.axvline(gamma.mean(),color ='r', linewidth='2')
plt.axvline(gamma.mean()+gamma.std(), color='y', linestyle='dashed')
plt.axvline(gamma.mean()-gamma.std(), color='y', linestyle='dashed')
plt.show()
```

```python
# Poisson Distribution


# lambda = 5
poisson = np.random.poisson(5, 1000)
print('The Mean is :{} , and the STD is :{}'.format(poisson.mean(), poisson.std()))
plt.hist(poisson)
plt.axvline(poisson.mean(),color ='r', linewidth='2')
plt.axvline(poisson.mean()+poisson.std(), color='y', linestyle='dashed')
plt.axvline(poisson.mean()-poisson.std(), color='y', linestyle='dashed')
plt.show() 
```

```python
# Bernoulli Distribution


# p =0.9
bernoulli= np.random.binomial(1, .9, 1000)
print('The Mean is :{} , and the STD is :{}'.format(bernoulli.mean(), bernoulli.std()))
plt.hist(bernoulli)
plt.axvline(bernoulli.mean(),color ='r', linewidth='2')
plt.axvline(bernoulli.mean()+bernoulli.std(), color='y', linestyle='dashed')
plt.axvline(bernoulli.mean()-bernoulli.std(), color='y', linestyle='dashed')
plt.show() 
```

```python
# Normal Distribution



# mean = 5 , std = 2
normal = np.random.normal(5,2, 1000)
print('The Mean is :{} , and the STD is :{}'.format(normal.mean(), normal.std()))
plt.hist(normal)
plt.axvline(normal.mean(),color ='r', linewidth='2')
plt.axvline(normal.mean()+normal.std(), color='y', linestyle='dashed')
plt.axvline(normal.mean()-normal.std(), color='y', linestyle='dashed')
plt.show() 
```

```python
norm1 = np.random.normal(5,.5,100)
norm2 = np.random.normal(10 , 1, 100)

comb = norm1 + norm2
print('The Mean is :{} , and the STD is :{}'.format(comb.mean(), comb.std()))
plt.hist(comb)
plt.axvline(comb.mean() , color='red' , linewidth='2')
plt.axvline(comb.mean() + comb.std(), color='orange')
plt.axvline(comb.mean() - comb.std(), color='orange')
plt.show()
print('\t\tGraph is skewed')
plt.plot(norm1 , comb , 'o')
plt.title('GG Plot(Norm1 compared to Combined)')
plt.show()
print('Combined array is no longer normally distributed')
```
